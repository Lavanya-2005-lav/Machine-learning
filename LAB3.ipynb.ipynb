```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```
df = pd.read_csv("canada_per_capita_income.csv")

# Check data
print(df.head())

# Prepare data
X = df[['year']]
y = df['per capita income (US$)']

# Create model
model = LinearRegression()
model.fit(X, y)

# Predict for year 2020
prediction = model.predict([[2020]])
print("Predicted Per Capita Income in 2020:", prediction[0])

# Plot
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.title("Canada Per Capita Income Prediction")
plt.show()
```

```
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv("salary.csv")

# Remove missing values
df = df.dropna()

X = df[['YearsExperience']]
y = df['Salary']

model = LinearRegression()
model.fit(X, y)

salary_12 = model.predict([[12]])
print("Predicted Salary for 12 years experience:", salary_12[0])
```

```
df = pd.read_csv("hiring.csv")

# Convert words to numbers
df['experience'] = df['experience'].replace({
    'zero':0,
    'one':1,
    'two':2,
    'three':3,
    'four':4,
    'five':5,
    'six':6,
    'seven':7,
    'eight':8,
    'nine':9,
    'ten':10,
    'eleven':11,
    'twelve':12
})

# Fill missing values if any
df.fillna(0, inplace=True)

print(df.head())
from sklearn.linear_model import LinearRegression

X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

model = LinearRegression()
model.fit(X, y)

# Predictions
print("2 yr exp, 9 test, 6 interview:",
      model.predict([[2,9,6]])[0])

print("12 yr exp, 10 test, 10 interview:",
      model.predict([[12,10,10]])[0])
```

```
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv("1000_Companies.csv")

# One-hot encoding for State
df = pd.get_dummies(df, columns=['State'], drop_first=True)

X = df.drop('Profit', axis=1)
y = df['Profit']

model = LinearRegression()
model.fit(X, y)

# Example prediction
prediction = model.predict([[91694.48, 515841.3, 11931.24, 1, 0]])
print("Predicted Profit:", prediction[0])
```
